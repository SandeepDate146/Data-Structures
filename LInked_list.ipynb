```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def add_to_end(self, value):
        ref = Node(value)
        if self.head == None:
            self.head = ref
        else:
            self.tail.next = ref
        self.tail = ref
    
    def add_to_front(self, value):
        ref = Node(value)
        if self.head != None:
            ref.next = self.head
        self.head = ref
        
    def show(self):
        current = self.head
        while current != None:
            print(current.value, end= " ")
            current = current.next
```

```python
L = LinkedList()
```

```python
L.add_to_end(5)
L.add_to_end(7)
L.add_to_front(3)
L.add_to_end(9)
L.show()
```
